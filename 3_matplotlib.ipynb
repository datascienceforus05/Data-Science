*Label In Graph*

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x= [1,5,4]
y= [23,49,83]
x2= [ 3,7,3]
y2= [5,8,9]
plt.plot(x,y, label="First")
plt.plot(x2,y2, label="Second")
plt.legend()
plt.show()
```

*Line Width*

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x= [1,5,4]
y= [23,49,83]
x2= [ 3,7,3]
y2= [5,8,9]
plt.plot(x,y, label="First" ,linewidth =4)
plt.plot(x2,y2, label="Second")
plt.legend()
plt.show()
```

*Plot Scatter*

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x= [1,5,4]
y= [23,49,83]
x2= [ 3,7,3]
y2= [5,8,9]
plt.scatter(x,y, label="First" ,linewidth =4)
plt.scatter(x2,y2, label="Second")
plt.legend()
plt.show()
```
