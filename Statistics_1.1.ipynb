**HYPOTHESIS TESTING** :- It is a part of statistical analysis, where we test the assumptions made regarding a population parameter. It is generally used when we were to compare a single group with an external standard and two or more groups with each other.

**Terminology Used In Hypothesis Testing**

*â¢Null Hypothesis* : Null Hypothesis is a Statistical theory that suggests there us no Statistical significance exists between the populations. It is denoted by H0 and read as H-naught.

*â¢Alternative Hypothesis* : An Alternative hypothesis suggests there is a significant difference between the population parameters. It could be greater or smaller. Basically, it is the contrast of the Null Hypothesis. It is denoted by Ha or H1.

*Hypothesis Testing* :-

â¢ Z-Test:- (Data --> In the average (Sample data))

â¢ T-Test:- (Data --> In the average (Sample data))

â¢ Chi-Square Test:- ( Tell Perfect or not , Tell the relation is available or not )

**STEPS OF HYPOTHESIS TESTING**

â¢ State null(H0) and alternative      (H1) hypothesis

â¢ Choose level or significance(a)

â¢ Find critical values

â¢ Find test statistic

â¢ Draw your conclusion

_________________________________

*â¢ State null(H0) and Alternative (H1) hypothesis:*

H0 must always contain equality (=). H1 always contains difference ( â  ,> ,<).

*â¢ Choose level or significance(a):*

Denoted by alpha or Î±. It is a fixed Probability of wrongly rejected a True Null Hypothesis.
(Î± value depends on Domain)

*â¢ Find Test Statistics:*

Z-Test :- When we have Mean of Population data, Standard Division and Sample data size > 30, Sample data mean.

T-Test :- When we have Population data mean, Standard division of sample data, sample data size < 30, sample data mean.

Chi-Square Test:- Check Goodness, Relationship between two variables.

*â¢ Find critical values:

Find the value of test ( Z-Test value, T-Test value )

*â¢Draw your conclusion:*

Then come to a conclusion. ( H0 or H1 is right? )

**Z-Test** :-

*Hypothesis testing (Z-Test)* --->>>

â¢Example 1:

A teacher vlaims that the mean score of students in his class is greater than 82 with a standard deviation of 20. If a sample of 81 students was selected with a mean score of 90.

[Î¼(mean) =82, Ï(standard deviation) = 20, n(sample data) = 81, xÌ(mean of sample data) = 90]

**Z-Test= xÌ-Î¼/(Ï/ân)**

__________________________________
Find H0 and H1 ---

H0 = Î¼ â  82

H1 = Î¼ > 82

Î± = 0.05 (see copy)

(Google :- Z score table..  Go to.. https://www.z-table.com/ !! Search 0.95 in table )

See corresponding row(1.6) and column (0.04) and add them ... Add them (1.64)

Z- value = 1.64

Z-calculation (Z-Test formula) = 90-82/(20/â81) = 3.6 (Z-Calculated)

*This is in H1 area, so the teacher is said true (Î¼ > 82)*

**In code** --->>>>

```python
import scipy.stats as st
st.norm.ppf(0.95) #This is calculate Z-value , without using z- table
```

```python
x = 90
u = 82 #Î¼
p_std = 20  #Ï
n = 81
ap = 0.05 #Î±
```

```python

import numpy as np
z_cal = (x-u)/(p_std/np.sqrt(n))
z_cal #this calculates the Z
```

```python
z_table = st.norm.ppf(1-ap)
z_table
```

```python
if z_table < z_cal :
    print("H1 is right")
else :
    print("H0 is right")
```
