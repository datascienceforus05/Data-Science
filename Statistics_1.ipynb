Mode :- Most Frequently Data (Repet Data) (Frequency = time of repetition) (**It used on Categorical Data**)

```python
df["Fare"].mode() #This return "Pandas core series" data so we have to change this by  using. [0]
```

```python
mo = df["Fare"].mode()[0] #It gives the mode
mo
```

```python
#Check frequency

df["Fare"].value_counts()
```

```python
sns.histplot(x= "Fare", data = df,bins = [i for i in range (0,81,10)])
plt.plot([mo for i in range (0,300)], [i for i in range (0,300)], c = "red")

plt.show()
```

Add Mean, Median, Mode in one Graph

```python
sns.histplot(x= "Age", data = df,bins = [i for i in range (0,81,10)])
plt.plot([mn for i in range (0,300)], [i for i in range (0,300)], c = "red", label ="Mean")
plt.plot([md_1 for i in range (0,300)], [i for i in range (0,300)], c = "blue", label = "Median")
plt.plot([mo for i in range (0,300)], [i for i in range (0,300)], c = "green", label = "Mode")
plt.legend()
plt.show()
```

Mean Absolute Division :- Tells the Average distance between each data point at the mean. It gives us an Idea about the Variability of a dataset

Formula :- (In copy)

```python
sec_a = np.array([75,65,73,68,72,67])
sec_b = np.array([90,47,43,96,93,51])
no = np.array ([1,2,3,4,5,6])
```

```python
mean_1 = np.mean(sec_b)
```

```python

plt.figure(figsize = (10,3))
plt.scatter(sec_a,no, label = "sec_a")
plt.scatter(sec_b,no, c= "red" , label = "sec_b")
plt.plot([70,70,70,70,70,70],no,c= "green", label = "Mean")
plt.show()
```

```python
#Using Mean Absolute division

mad_a = np.sum(np.abs(sec_a - mean_1))/len(sec_a)
mad_b = np.sum(np.abs(sec_b - mean_1))/len(sec_b)

mad_a,mad_b
```

**mad_b** Data spredness is too much so then we use **mad_a**
